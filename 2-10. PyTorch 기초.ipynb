# AutoGrad

```python
x = torch.ones(2, 2, requires_grad=True)
x
```

```python
y = x + 2
y
```

```python
print(y.grad_fn)
```

```python
z = y * y * 3
out = z.mean()
```

```python
print(z, out)
```

```python
out.backward()
```

```python
print(x.grad)
```

```python
print(x.requires_grad)
print((x ** 2).requires_grad)
```

```python
with torch.no_grad():
    print((x ** 2).requires_grad)
```

```python
w = torch.tensor(1.0, requires_grad = True) #required_grad = True는 Autograd가 계산할때 w에 대한 미분값을 w.grad에 저장한다.
a = w*3
```

```python
l = a**2  # l = a제곱 = (3w)제곱 
```

```python
l.backward()
print('l을 w로 미분한 값은 {}'.format(w.grad))
```
